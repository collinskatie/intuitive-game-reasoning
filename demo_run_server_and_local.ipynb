# Running on cluster

## Just think experiment

### Add states to simulate
First, open the `models/just_think/recover_missing.py` file, set `model_name` to `heuristics`, `heuristic_search`, `mcts`, or `uni_random`, and set the date to the current date

```python
!python models/just_think/recover_missing.py
# Returns the number of states to simulate
```

### Run simulations
First, open the `models/just_think/run_models.py` file, set `model_name` to `heuristics`, `heuristic_search`, `mcts`, or `uni_random`, and set the date to the current date

Run SLURM script for running simulations in cluster. This will generate many files that contain data about the simulations.

```python
!bash models/just_think/shell_scripts/run_models_shell.sh
```

### Merge files
First, open the `models/just_think/merge_files/merge_files.py` file, set `model_name` to `heuristics`, `heuristic_search`, `mcts`, or `uni_random`, and set the date to the current date

After running this script, you can find your simulations in `models/just_think/merged_results/<date>/<model-name>/{}.txt`.

```python
!bash models/just_think/shell_scripts/merge_files.sh
```
